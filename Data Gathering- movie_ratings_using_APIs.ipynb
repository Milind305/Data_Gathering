# APIs Tutorial

Importing libraries (inbuilt library in anaconda)

```python
import requests
import pandas as pd
```

Importing the web api in json format

```python
response=requests.get('https://api.themoviedb.org/3/movie/popular?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1').json()
response
```

```python
popularity=[]
vote_count=[]
original_lang=[]
title=[]
vote_avg=[]
overview=[]
release_date=[]
for i in response['results']:
    popularity.append(i['popularity'])
    vote_count.append(i['vote_count'])
    original_lang.append(i['original_language'])
    title.append(i['title'])
    vote_avg.append(i['vote_average'])
    overview.append(i['overview'])
    release_date.append(i['release_date'])
    
d= {'title':title, 'overview': overview, 'original_lang': original_lang, 'release_date': release_date, 'popularity': popularity, 'vote_count':vote_count, 'vote_avg':vote_avg}
    
df=pd.DataFrame(d)
df
```

```python
df1=pd.DataFrame()
for j in range (1,500):
    response=requests.get('https://api.themoviedb.org/3/movie/popular?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}' .format(j)).json()
    popularity=[]
    vote_count=[]
    original_lang=[]
    title=[]
    vote_avg=[]
    overview=[]
    
    for i in response['results']:
        popularity.append(i['popularity'])
        vote_count.append(i['vote_count'])
        original_lang.append(i['original_language'])
        title.append(i['title'])
        vote_avg.append(i['vote_average'])
        overview.append(i['overview'])
    
    d= {'title':title, 'overview': overview, 'original_lang': original_lang, 'popularity': popularity, 'vote_count':vote_count, 'vote_avg':vote_avg}
    
    df2=pd.DataFrame(d)
    
    df1=df1.append(df, ignore_index=True)
```
