##### Loading Libraries

```python
import requests
import pandas as pd
```

Use of get function

```python
response=requests.get('https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest?CMC_PRO_API_KEY=4d7c9d77-f0d4-4187-b55c-4ef3f5206cfe').json()
response
```

```python
#empty list for the features to be captured
curr_name=[]
curr_price=[]
curr_change_1h=[]
curr_change_24h=[]
curr_change_7d=[]
curr_change_30d=[]

#loop for appending the features into the list from the JSON
for i in response['data']:
    curr_name.append(i['name'])
    curr_price.append(i['quote']['USD']['price'])
    curr_change_1h.append(i['quote']['USD']['percent_change_1h'])
    curr_change_24h.append(i['quote']['USD']['percent_change_24h'])
    curr_change_7d.append(i['quote']['USD']['percent_change_7d'])
    curr_change_30d.append(i['quote']['USD']['percent_change_30d'])
    
    d={'curr_name':curr_name, 'curr_price':curr_price, 
       'curr_change_1h':curr_change_1h, 'curr_change_24h':curr_change_24h,
       'curr_change_7d':curr_change_7d, 'curr_change_30d':curr_change_30d}

# storing the data into the dataframe
df3=pd.DataFrame(d)

    
    
```
